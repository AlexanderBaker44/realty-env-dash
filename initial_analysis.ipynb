```python
import pandas as pd
```

```python
df_effective = pd.read_excel('data/dallas_apartment_data_rent_sf.xlsx',sheet_name = 'Effective $ Per SF')
df_effective.columns = [str(i).split(' ')[0] for i in df_effective.columns]
df_effective['MSA/Submarket'] = [f'{i}_effective' for i in df_effective['MSA/Submarket']]
df_effective.head()
```

```python
df_market = pd.read_excel('data/dallas_apartment_data_rent_sf.xlsx',sheet_name = 'Market $ Per SF')
df_market.columns = [str(i).split(' ')[0] for i in df_market.columns]
df_market['MSA/Submarket'] = [f'{i}_market' for i in df_market['MSA/Submarket']]
df_market.head()
```

```python
df_full_market = pd.concat([df_market,df_effective])
df_full_market
```

```python
df_full_t = df_full_market.T
df_full_t.columns = df_full_market['MSA/Submarket']
df_full_t.drop('MSA/Submarket',axis=0,inplace = True)
df_full_t
```

```python
df_full_t.index = pd.to_datetime(df_full_t.index)
```

```python
import matplotlib.pyplot as plt
plt.plot(df_full_t['Dallas/Fort Worth Overall_effective'])
plt.plot(df_full_t['Dallas/Fort Worth Overall_market'])
```

```python
df_supply = pd.read_excel('data/dallas_apartment_data_rent_sf.xlsx',sheet_name = 'Supply')
df_supply = df_supply[df_supply['MSA/Submarket']!='Dallas/Fort Worth Overall']
df_supply.head()
#print(df_supply.columns)
```

```python
df_supply
```

```python
df_supply.sort_values('Total Units ',ascending =False,inplace=True)
```

```python
fig,ax_l= plt.subplots(1,1)
plt.bar(x = df_supply['MSA/Submarket'],height = df_supply['Total Units '])
plt.xticks(rotation=90)
plt.show()
```

```python
df_supply['delivery_differences'] = df_supply['2024 Deliveries (YTD)'] - df_supply['2023 Deliveries']
```

```python
df_supply.sort_values('delivery_differences',ascending =False,inplace=True)
df_supply.plot(x = 'MSA/Submarket',y = 'delivery_differences',kind = 'bar')
```

```python
df_supply.sort_values('2024 Net Absorption (YTD)',ascending =False,inplace=True)
df_supply.plot(x = 'MSA/Submarket',y = '2024 Net Absorption (YTD)',kind = 'bar')
```
