```python
df['6 Year Change'] = df['December 2010'] - df['December 2016']
```

```python
for i in df.columns:
    if 'December 20' in i and '10' not in i:
        split_list = i.split(' ')
        df[f'{split_list[0]}_{split_list[1]}_{int(split_list[1])-1}'] = df[f'{split_list[0]} {split_list[1]}']-df[f'{split_list[0]} {int(split_list[1])-1}']
        
```

```python
col_list = [i for i in df.columns if '_20' in i]
print(col_list)
average_momentum = df[col_list].fillna(0).sum(axis=1)
average_momentum
df['average_momentum'] = average_momentum/len(col_list)
```
